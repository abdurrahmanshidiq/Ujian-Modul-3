## #3

```python
from sklearn.model_selection import train_test_split
```

```python
scaled_nba.head()
```

```python
X = scaled_nba.drop('best_player', axis=1)
y = scaled_nba['best_player']
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.78, test_size=.22, random_state=101)
```

```python
print(f'X_train = {len(X_train)}')
print(f'X_test = {len(X_test)}')
print('-'*15)
print(f'y_train = {len(y_train)}')
print(f'y_test = {len(y_test)}')
```

## #4

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Hyper Parameter Tuning
from sklearn.model_selection import RandomizedSearchCV
```

```python
rfc = RandomForestClassifier()
```

```python
param_random = {
    'bootstrap': [True, False],
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [int(x) for x in np.arange(10,120,20)],
    'min_samples_split' : [int(x) for x in np.arange(2,32,2)],
    'n_estimators' : [int(x) for x in np.linspace(start=100, stop=1000, num=10)]
}

rfc_random = RandomizedSearchCV(estimator= rfc,
                              param_distributions= param_random,
                              n_iter=10, cv=5, n_jobs=-1, scoring='accuracy')
```

```python
rfc_random.fit(X_train,y_train)
```

```python
rfc_random.best_params_
```

```python
rfc_rand = rfc_random.best_estimator_
rfc_rand
```

### KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn
```

```python
param_knn = {
    'leaf_size': [int(x) for x in np.arange(10,110,20)],
    'weights' : ['uniform', 'distance'],
    'n_neighbors' : [int(x) for x in range(1,80)],
}

knn_random = RandomizedSearchCV(estimator= knn,
                              param_distributions= param_knn,
                              n_iter=10, cv=5, n_jobs=-1, scoring='accuracy')
```

```python
knn_random.fit(X_train, y_train)
```

```python
knn_random.best_params_
```

```python
knn_rand = rfc_random.best_estimator_
knn_rand
```

### Logistic Regression

### ROC AUC all Models

```python
prob_rfc = rand_proba[:,1]
fpr_rfc, tpr_rfc, threshold_rfc = roc_curve(y_test, prob_rfc)
roc_auc_rfc = auc(fpr_rfc, tpr_rfc)

prob_knn = knn_proba[:,1]
fpr_knn, tpr_knn, threshold_knn = roc_curve(y_test, prob_knn)
roc_auc_knn = auc(fpr_knn, tpr_knn)

prob_logreg = logreg_proba[:,1]
fpr_logreg, tpr_logreg, threshold_logreg = roc_curve(y_test, prob_logreg)
roc_auc_logreg = auc(fpr_logreg, tpr_logreg)
```

```python
plt.figure(figsize=(20,7))

plt.subplot(131)
plt.title('ROC AUC Random Forest Classifier')
plt.plot(fpr_rfc, tpr_rfc, 'green', label='AUC RFC = {}'.format(round(roc_auc_rfc,2)))
plt.plot([0,1],[0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.fill_between(fpr_rfc,tpr_rfc, 0, facecolor='azure', alpha=0.5)
plt.legend(loc = 'lower right')

plt.subplot(132)
plt.title('ROC AUC KNN')
plt.plot(fpr_knn, tpr_knn, 'green', label='AUC KNN = {}'.format(round(roc_auc_knn,2)))
plt.plot([0,1],[0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.fill_between(fpr_knn,tpr_knn, 0, facecolor='azure', alpha=0.5)
plt.legend(loc = 'lower right')

plt.subplot(133)
plt.title('ROC AUC Logistic Regression')
plt.plot(fpr_logreg, tpr_logreg, 'green', label='AUC Logistic = {}'.format(round(roc_auc_logreg,2)))
plt.plot([0,1],[0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.fill_between(fpr_logreg,tpr_logreg, 0, facecolor='azure', alpha=0.5)
plt.legend(loc = 'lower right')
```

- Model Random Forest Classifier & KNN memiliki skor AUC 1.0 yang mana AUC menutupi seluruh bidang plot.
- Model Logistic Regression memiliki skor ACU 0.98

- Model Random Forest Classifier & KNN adalah yg terbaik
